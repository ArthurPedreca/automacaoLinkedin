```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time 

driver = webdriver.Chrome()
driver.get("https://www.linkedin.com/feed/")
driver.maximize_window()

#Login 

email_elem = driver.find_element(By.XPATH, "/html/body/div/main/div[2]/div[1]/form/div[1]/input")
email_elem.send_keys("arthurpedreca@gmail.com")

senha_elem = driver.find_element(By.XPATH, "/html/body/div/main/div[2]/div[1]/form/div[2]/input")
senha_elem.send_keys("SenhaMuitoComplexa123")

entrar_elem = driver.find_element(By.XPATH, "/html/body/div/main/div[2]/div[1]/form/div[3]/button")
entrar_elem.click()


#Fazer o filtro


pesquisa_elem = WebDriverWait(driver, 30).until(EC.visibility_of_element_located((By.XPATH, '/html/body/div[5]/header/div/div/div/div[1]/input')))
pesquisa_elem.send_keys("Suporte Administrativo" + Keys.ENTER)

pessoas_elem = WebDriverWait(driver, 30).until(EC.visibility_of_element_located((By.XPATH, '/html/body/div[5]/div[3]/div[2]/section/div/nav/div/ul/li[3]/button')))
pessoas_elem.click()

todos_filtros_elem = WebDriverWait(driver, 30).until(EC.visibility_of_element_located((By.XPATH, '/html/body/div[5]/div[3]/div[2]/section/div/nav/div/div/div/button')))
todos_filtros_elem.click()

saopaulo_sp = WebDriverWait(driver, 30).until(EC.visibility_of_element_located((By.XPATH, '/html/body/div[3]/div/div/div[2]/ul/li[4]/fieldset/div/ul/li[4]/label')))
saopaulo_sp.click()

operacoes_elem = WebDriverWait(driver, 30).until(EC.visibility_of_element_located((By.XPATH, '/html/body/div[3]/div/div/div[2]/ul/li[11]/fieldset/div/ul/li[1]/label')))
operacoes_elem.click()

resultados_elem = WebDriverWait(driver, 30).until(EC.visibility_of_element_located((By.XPATH, '/html/body/div[3]/div/div/div[3]/div/button[2]')))
resultados_elem.click()

quantidade_resultados_elem = WebDriverWait(driver, 30).until(EC.visibility_of_element_located((By.XPATH, '/html/body/div[5]/div[3]/div[2]/div/div[1]/main/div/div/div[1]/h2/div')))
print(quantidade_resultados_elem.text)

for i in range(3):
    try:
        lead_elem = WebDriverWait(driver, 30).until(
            EC.visibility_of_element_located((By.XPATH, f'/html/body/div[5]/div[3]/div[2]/div/div[1]/main/div/div/div[2]/div/ul/li[{i + 1}]/div/div/div/div[3]'))
        )
        filhos = lead_elem.find_elements(By.XPATH, "./*")

        print(f"Verificando Lead número: {i+1}")
        if filhos:
            nome_lead = WebDriverWait(driver, 30).until(
                EC.visibility_of_element_located((By.XPATH, f"/html/body/div[5]/div[3]/div[2]/div/div[1]/main/div/div/div[2]/div/ul/li[{i + 1}]/div/div/div/div[2]/div[1]/div[1]/div/span[1]/span/a/span/span[1]"))
            )
            print(nome_lead.text)

            # Tentar localizar o botão com diferentes seletores
            try:
                button_element = driver.find_element(By.XPATH, f'/html/body/div[5]/div[3]/div[2]/div/div[1]/main/div/div/div[2]/div/ul/li[{i + 1}]/div/div/div/div[3]/div/button')
            except:
                button_element = driver.find_element(By.XPATH, f'//*[@class="artdeco-button artdeco-button--2 artdeco-button--secondary ember-view"]')
            
            aria_label = button_element.get_attribute("aria-label")

            # Verifique o conteúdo do atributo 'aria-label' e realize as ações correspondentes
            if aria_label.startswith("Seguir "):
                # Clique no botão "Seguir"
                button_element.click()
                print("Clicou no botão Seguir")

            elif aria_label.startswith("Enviar mensagem para"):
                # Clique no botão "Enviar mensagem"
                button_element.click()
                print("Clicou no botão Enviar mensagem")
                time.sleep(3)
                fechar_elem = driver.find_element(By.XPATH, '/html/body/div[5]/div[4]/aside[1]/div[2]/div[1]/header/div[3]/button[2]')
                fechar_elem.click()

            elif aria_label.startswith("Convidar ") and "para se conectar" in aria_label:
                # Clique no botão "Convidar para se conectar"
                button_element.click()
                print("Clicou no botão Convidar para se conectar")
                time.sleep(3)
                fechar_elem = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/button')
                fechar_elem.click()
                # Adicione aqui a ação específica para "Convidar para se conectar")
                
    except Exception as e:
        print(f"Ocorreu um erro: {e}")

```
